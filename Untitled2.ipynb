```python
import json
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
from sklearn.tree import export_graphviz
import graphviz
import matplotlib.pyplot as plt

import pydotplus
from IPython.display import Image
from sklearn import tree

# Membaca data dari file JSON
ed = "/content/drive/MyDrive/Colab Notebooks/csvjson.json"
with open(ed, 'r') as f:
    data = json.load(f)

# Mengubah data menjadi DataFrame
df = pd.DataFrame(data)
df
```

```python
# Mengubah data string ke dalam bentuk numerik menggunakan LabelEncoder
encoder = LabelEncoder()
for column in df.columns:
    df[column] = encoder.fit_transform(df[column])

# Memisahkan feature dan target
X = df.drop('Status', axis=1)
y = df['Status']
```

```python
# Memisahkan feature dan target
X = df.drop('Status', axis=1)
y = df['Status']
```

```python
df
```

```python
#@title Default title text
# Membuat objek DecisionTreeClassifier
clf = DecisionTreeClassifier(criterion='entropy')
# Melatih model dengan data
clf.fit(X, y)

# Membuat prediksi dengan model yang telah dilatih
predictions = clf.predict(X)

# Menghitung akurasi model
accuracy = accuracy_score(y, predictions)
print(f'Akurasi model: {accuracy:.2f}')
```

```python
# Menampilkan pohon keputusan
plt.figure(figsize=(20, 10))
plot_tree(clf, filled=True, feature_names=X.columns)
plt.show()
```

```python

# Menampilkan tabel hasil pohon keputusan dalam bentuk JSON
tree_dict = {'max_depth': clf.tree_.max_depth,
             'node_count': clf.tree_.node_count,
             'capacity': clf.tree_.capacity,
             'children_left': clf.tree_.children_left.tolist(),
             'children_right': clf.tree_.children_right.tolist(),
             'feature': clf.tree_.feature.tolist(),
             'threshold': clf.tree_.threshold.tolist(),
             'value': clf.tree_.value.tolist()}

print(json.dumps(tree_dict))
```
